## Slide 9 - Distribution of training data

This plot shows the distribution of two sets of recorded data: relax (eyes closed) and focus (eyes open). The dot plot is a nice, simple (1D) way of displaying the two classes on a single axis.

I didn't find a standard implementation of the dot plot for matplotlib, and so I made my own implementation (probably not super robust or efficient).

```python
# Generate fake relax/focus data
nb_points = 10*10
relax_data = np.random.normal(0.01, 0.01, size=(nb_points,))
focus_data = np.random.normal(0.03, 0.01, size=(nb_points,))

utils.dot_plot(x=np.concatenate((relax_data, focus_data)), 
         labels=np.concatenate((np.zeros((nb_points,)), np.ones((nb_points,)))),
         step=4)
```

## Slide 7 - PSD visualization

A simple visualization of the PSD of a 1-s window of raw EEG.

```python
# Generate fake PSD data
f = np.arange(0, 110, 1) # one-second windows = 1-Hz bins
psd = 10*np.log10(1./f)

utils.psd_with_bands_plot(f, psd)
```
